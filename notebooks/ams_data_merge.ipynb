# Ams data merge

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pvlib
```

## Load data

```python
fname_ams_weather = '../data/nl/ams_data_weather_processed.csv'
fname_ams_energy = '../data/nl/ams_data_energy.csv'
df_ams_weather, df_ams_energy = [pd.read_csv(fname) for fname in [
    fname_ams_weather, fname_ams_energy]]
```

```python
fname_ams_pv = '../data/nl/ams_data_pv.csv'
df_ams_pv = pd.read_csv(fname_ams_pv)
df_ams_weather.head(30)
```

## Add weather data to ams energy data

```python
df_ams_weather.columns
```

```python
# print the shape of df_ams_weather, df_ams_energy and df_ams, using one fstring
print(
    f'Shape of df_ams_weather: {df_ams_weather.shape} \nShape of df_ams_energy is {df_ams_energy.shape}')
```

```python
# Merge weather data and energy data into one dataframe called df_ams
# based on df_ams_energy
# add irradiance, temperature, could_cover, wind_speed, and precipitation, and their normalized features it, in total 10 features
# then obtain df_ams, which should have a shape of (16960,88+10)
# pay attention to the duplicate columns of time horizon:['HoD', 'dow', 'doy', 'month', 'year']
df_ams = pd.merge(df_ams_weather, df_ams_energy, on=None, how='inner')
# the parameter "on" was set to None, which means that the merge will be based on the common columns in the two dataframes
```

```python
df_ams_pv.columns
```

```python
# do the same for the dataframes that stores the PV, consumption and net load separately
df_ams_pv = pd.merge(df_ams_weather, df_ams_pv, on=None, how='inner')
df_ams_con = pd.merge(df_ams_weather, df_ams_energy, on=None, how='inner')
df_ams_net = pd.merge(df_ams_weather, df_ams_energy, on=None, how='inner')
```

```python
df_ams_pv.columns
```

```python
# print the shape of df_ams_weather, df_ams_energy and df_ams, using one fstring
print(
    f'Shape of df_ams_weather: {df_ams_weather.shape} \nShape of df_ams_energy is {df_ams_energy.shape} \nShape of df_ams {df_ams.shape}')
```

```python
# get the different columns between df_ams and df_ams_energy
# and print the different columns
diff_cols = set(df_ams.columns)-set(df_ams_energy.columns)
print(diff_cols)
```
